```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split as TSSplit
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
train_data = pd.read_csv("./train.csv")
X = train_data.iloc[:,0:13]
y = train_data.iloc[:,-1]
X.drop(['PassengerId','Name'],axis = 'columns',inplace = True)
X_train,X_valid,y_train,y_valid = TSSplit(X,y)
X.columns
```

```python
X_train_ed = X_train.copy()
one_hot = OneHotEncoder(sparse_output = False)
encoded_array = one_hot.fit_transform(X_train_ed[["HomePlanet","Destination"]])
encoded_labels = one_hot.categories_
encoded_labels = np.array(encoded_labels).ravel()
```

```python
encoded_features = pd.DataFrame(encoded_array,columns = encoded_labels)
t = encoded_features.columns
encoded_features.drop(t[3],axis = 1,inplace = True)
encoded_features.head()
```

```python
X_train_ed.shape , encoded_features.shape
```

```python
# Completed encoding categorical Features
X_train_ed.reset_index(drop=True, inplace=True)
encoded_features.reset_index(drop=True, inplace=True)
X_train_ed = pd.concat([encoded_features,X_train_ed],axis = 1)
X_train_ed.drop(['HomePlanet','Destination'],axis = 1,inplace = True)
X_train_ed.head()
```

```python
#Converting Cabins to categorical data:
cabins = X_train_ed.Cabin
cabins.head()
cate = []
freq = {}
for count , i in enumerate(cabins):
    i = str(i)
    cabin = i[0]
    if cabin in freq:
        freq[cabin] +=1
    else:
        freq[cabin] = 1
    cate.append(cabin)
print(freq)
```

```python
cabins = pd.Series(cate)
cabins.head()
X_train_ed.Cabin = cabins
X_train_ed.head()
```

```python
OHE = OneHotEncoder(sparse_output = False)
encoded_array = OHE.fit_transform(X_train_ed[["Cabin"]])
encoded_features = OHE.categories_
encoded_df  = pd.DataFrame(encoded_array,columns = encoded_features)
encoded_df.drop('n',axis = 1,inplace = True)
```

```python
X_train_ed.reset_index(drop=True, inplace=True)
encoded_df.reset_index(drop=True, inplace=True)
X_train_ed = pd.concat([encoded_df,X_train_ed],axis = 1)
X_train_ed.drop(['Cabin'],axis = 1,inplace = True)
X_train_ed.head()
```

```python
X_train_total = X_train_processed.copy()
X_train_total = X_train_total._append(X_valid_processed)
X_train_total.shape
```

```python
from xgboost import XGBClassifier
from datetime import datetime
def timer(start_time=None):
    if not start_time:
        start_time = datetime.now()
        return start_time
    elif start_time:
        thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
        tmin, tsec = divmod(temp_sec, 60)
        print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))
```

```python
xgb = XGBClassifier(learning_rate=0.02, objective='binary:logistic')
params = {
    'n_estimators' : [100,200,300,400,500,600,700,800,900,1000],
    'max_depth' : [3,4,5,6,7]
}

# preds = bst.predict(X_test)
```

```python
X_train_total.CryoSleep = X_train_total.CryoSleep.astype('bool')
X_train_total.CryoSleep.head()
```

```python
X_train_total.VIP = X_train_total.VIP.astype('bool')
X_train_total.VIP.head()
X_train_total.dtypes
```

```python
from sklearn.model_selection import GridSearchCV
start_time = timer(None)
grid = GridSearchCV(estimator=xgb, param_grid=params, scoring='roc_auc', n_jobs=-1, cv=3, verbose=3 )
grid.fit(X_train_total,y)
timer(start_time)
print('\n All results:')
print(grid.cv_results_)
print('\n Best estimator:')
print(grid.best_estimator_)
print('\n Best score:')
print(grid.best_score_ * 2 - 1)
print('\n Best parameters:')
print(grid.best_params_)
results = pd.DataFrame(grid.cv_results_)
results.to_csv('xgb-grid-search-results-01.csv', index=False)

# y_test = grid.best_estimator_.predict_proba(test)
# results_df = pd.DataFrame(data={'id':test_df['id'], 'target':y_test[:,1]})
# results_df.to_csv('submission-grid-search-xgb-porto-01.csv', index=False)
```
